```python

import pandas as pd
part=pd.read_csv(r'C:\Users\schro\Desktop\Projects\HackHunger\MealPart-riv.csv', header=0, sep=', ', delimiter=',')
ins=pd.read_csv(r'C:\Users\schro\Desktop\Projects\HackHunger\FoodInsecurity.csv', header=0, sep=', ', delimiter=',')
ins2=pd.read_csv(r'C:\Users\schro\Desktop\Projects\HackHunger\FoodInsecurity2.csv', header=0, sep=', ', delimiter=',')

#too add in Municipalities:
#ins_tot=ins.append(ins2, ignore_index=True)
ins_tot= ins.rename(columns = {'Chicago Community Area':'Community Area'})

#fix riverdale so that both spreadsheets can match
part.loc[part['Community Area'] == 'Riverdale - Chicago', 'Community Area'] = "Riverdale"

#get uppercase and list of communities for map
ins_tot['Community_upper']=ins_tot['Community Area'].str.upper()
part['Community_upper']=part['Community Area'].str.upper()

#create ins for d3
ins_tot['insecurity']=ins_tot['Food Insecurity Rate']
ins_tot['insecurity'].replace(to_replace=r'%',value=r'', regex=True, inplace=True)
ins_tot.dropna(subset = ['insecurity'],inplace=True)
total_df=pd.merge(ins_tot,part,on='Community_upper',how='left')

#total_df.to_csv(r'C:\Users\schro\Desktop\Projects\HackHunger\total_.csv')
print total_df.info()
```

```python
#pull geocodes for all communities
import geocoder
n=0
total_df_geo=total_df
total_df_geo['geocode']=0
while n<77:
    district = total_df_geo['Community_upper'][n]
    g = geocoder.google(district+', Chicago')
    total_df_geo['geocode'][n]=g
    n+=1
```

```python
communities=[]
for comm in total_df_geo['Community_upper']:
        communities.append(comm)
print communities

#pull latitude and longitude for each geocode
total_df_geo['lat']=float(0)
total_df_geo['lon']=float(0)
print total_df_geo.info()
n=0
while n<77:
    if len(total_df_geo['geocode'][n].latlng) >1: 
        total_df_geo['lat'][n]=float(total_df_geo['geocode'][n].latlng[0])
        total_df_geo['lon'][n]=float(total_df_geo['geocode'][n].latlng[1])
        n+=1
    else:
        n+=1
print total_df_geo.head()

r=0
for row in total_df_geo['lat']:
    if row==0:
        total_df_geo.drop(r)
    r+=1
total_df_geo.to_csv(r'C:\Users\schro\Desktop\Projects\HackHunger\total_food_riv.csv')


```
